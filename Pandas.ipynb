# Operacões que modificam o dataframe original

```python
# resetando os rótulos do índice
'''
Ao aplicarmos o método reset_index criamos um novo índice com
rótulo numérico e o rótulo anterior vira uma nova coluna no dataframe
'''
nome = ['Douglas','Daniela','Pedro','Maria','Eduardo','Ester']
idade = [45,7,65,64,42,37]
altura = [1.85,1.23,1.75,1.67,1.82,1.73]
peso = [70,22,87,64,96,68]
sexo = ['M','F','M','F','M','F']
df = pd.DataFrame(zip(nome,idade,altura,peso,sexo),
                  columns=['nome','idade','altura','peso','sexo'])
df = df.rename({0:'pai',1:'filha'}, axis='rows')
display(df)
df.reset_index(inplace=True)
display(df)
```

```python
# excluindo colunas
'''
Ao aplicarmos o método reset_index criamos um novo índice com
rótulo numérico e o rótulo anterior vira uma nova coluna no dataframe
'''
nome = ['Douglas','Daniela','Pedro','Maria','Eduardo','Ester']
idade = [45,7,65,64,42,37]
altura = [1.85,1.23,1.75,1.67,1.82,1.73]
peso = [70,22,87,64,96,68]
sexo = ['M','F','M','F','M','F']
df = pd.DataFrame(zip(nome,idade,altura,peso,sexo),
                  columns=['nome','idade','altura','peso','sexo'])
df = df.rename({0:'pai',1:'filha'}, axis='rows')
display(df)
df.reset_index(inplace=True)
display(df)
'''
Agora vamos excluir a coluna gerada automaticamente no momento
em que resetamos o índice
'''
df.drop('index',axis=1,inplace=True)
display(df)
```
